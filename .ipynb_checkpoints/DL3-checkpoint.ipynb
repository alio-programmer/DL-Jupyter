Build the Image classification model by dividing the model into following 4 stages: 
a. Loading and preprocessing the image data 
b. Defining the model’s architecture 
c. Training the model 
d. Estimating the model’s performance

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import seaborn as sns
```

```python
#importing the libraries
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from sklearn.metrics import classification_report
```

```python
from tensorflow.keras.layers import BatchNormalization, Dropout
from tensorflow.keras.optimizers import SGD
```

```python
#grabbing CIFAR10 dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
#showing images of mentioned categories
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer','dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10,10))
for i in range(10):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()
```

```python
model = models.Sequential()

# 1st Convolutional Layer
model.add(layers.Conv2D(64, (3, 3), strides=(1, 1), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))
model.add(BatchNormalization())

# 2nd Convolutional Layer
model.add(layers.Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))
model.add(BatchNormalization())


# 5th Convolutional Layer
model.add(layers.Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

# Flatten
model.add(layers.Flatten())

# Fully Connected Layer 1
model.add(layers.Dense(256, activation='relu'))
model.add(Dropout(0.5))

# Fully Connected Layer 2
model.add(layers.Dense(256, activation='relu'))
model.add(Dropout(0.5))

# Output Layer (adjust units for number of classes, e.g., 10 for CIFAR-10)
model.add(layers.Dense(10))

model.summary()
```

```python
#model compilation
sgd = SGD(0.1)
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
epochs = 10
h = model.fit(train_images, train_labels, epochs=epochs, validation_data=(test_images, test_labels), batch_size=128)
```

```python
plt.matshow(test_images[666])
```

```python
predictions = model.predict(test_images, batch_size=128)
```

```python
class_names[np.argmax(predictions[666])]
```

```python
y_predicted_labels = [np.argmax(i) for i in predictions]
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(test_labels, y_predicted_labels)
plt.figure(figsize = (10, 7))
sns.heatmap(cm, annot=True, fmt='d')
#this heatmap shows that our model suffered most in the classification of classes 2 and 3 i.e birds and cats
```

```python
model_eval = model.evaluate(test_images, test_labels)
print("Accuracy of model is:", model_eval[1], "%")
```
